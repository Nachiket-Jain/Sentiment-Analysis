```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv("E:/nachiket/RP/Dataset-SA.csv")
```

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
```

```python
print("Missing Values Before Cleaning:")
print(df.isnull().sum())
```

```python
sentiment_fill = {
    "positive": "this is a good product",
    "negative": "this is a bad product",
    "neutral": "average or okay product"
}

df.loc[df["Review"].isna(), "Review"] = df["Summary"]

df.loc[df["Summary"].isna(), "Summary"] = df["Review"]

df["Review"] = df.apply(lambda x: sentiment_fill.get(x["Sentiment"], "no review") if pd.isna(x["Review"]) or x["Review"] == "" else x["Review"], axis=1)
df["Summary"] = df.apply(lambda x: sentiment_fill.get(x["Sentiment"], "no summary") if pd.isna(x["Summary"]) or x["Summary"] == "" else x["Summary"], axis=1)
```

```python
def clean_text(text):
    if not isinstance(text, str):
        text = str(text)
    text = re.sub(r"http\S+", "", text)                 
    text = re.sub(r"www\S+", "", text)
    text = re.sub(r"@\w+|#\w+", "", text)                 
    text = re.sub(r"[^\w\s\u0900-\u097F]", "", text)        
    text = re.sub(r"\s+", " ", text).strip()              
    return text.lower()
```

```python
df["Review"] = df["Review"].apply(clean_text)
df["Summary"] = df["Summary"].apply(clean_text)
```

```python
df.drop_duplicates(subset=["Review", "Summary"], inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
print("Missing Values After Filling:")
print(df.isnull().sum())
```

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
```

```python
print("Sample Filled and Cleaned Data:")
print(df.sample(5, random_state=42)[["Review", "Summary", "Sentiment"]])
```

```python
save_path = r"E:\nachiket\RP\Dataset-SA.csv"

df.to_csv(save_path, index=False, encoding="utf-8-sig")

print(f"Cleaned dataset successfully saved at:\n{save_path}")
print(f"Total rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
import matplotlib.pyplot as plt

sentiment_counts = df["Sentiment"].value_counts()

print("Sentiment Distribution:")
print(sentiment_counts)

plt.figure(figsize=(6,4))
sentiment_counts.plot(kind="bar")
plt.title("Sentiment Distribution in Cleaned Dataset")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()

```
